## crawler


```python
def get_director(url):
    response = get(url)
    html_soup = BeautifulSoup(response.text, 'html.parser')
    director = html_soup.find('td', class_ = 'name').a.text
    
    return(director)
```

```python
# csv 파일로 저장
result = []
for i in range(len(url)):
    data = get_director(url["url"][i])
    result.append(data)
result = pd.DataFrame(result)
result.to_csv("director.csv", index=False)
```

```python
def get_actor(url):
    response = get(url)
    html_soup = BeautifulSoup(response.text, 'html.parser')
    
    actor_containers_odd = html_soup.find_all('tr', class_ = 'odd')
    actor_containers_even = html_soup.find_all('tr', class_ = 'even')
    actor_containers = actor_containers_odd + actor_containers_even
    
    li = []
    for i in range(len(actor_containers)):
        li.append(actor_containers[i].td.img['alt'])
    
    return(li)
```

```python
# csv 파일로 저장
result = []
for i in range(len(url)):
    data = get_actor(url["url"][i])
    result.append(data)
result = pd.DataFrame(result)
result.to_csv("actor.csv", index=False)
```

```python
def get_character(url):
    response = get(url)
    html_soup = BeautifulSoup(response.text, 'html.parser')
    
    characters = html_soup.find_all('td', class_ = 'character')
    
    li = []
    for i in range(len(characters)):
        try:
            li.append(characters[i].a.text)
        except AttributeError:
            pass
           
    return(li)
```

```python
# csv 파일로 저장
result = []
for i in range(len(url)):
    data = get_character(url["url"][i])
    result.append(data)
result = pd.DataFrame(result)
result.to_csv("character.csv", index=False)
```
