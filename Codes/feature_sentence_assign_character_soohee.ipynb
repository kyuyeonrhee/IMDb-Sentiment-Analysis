```python
import pandas as pd
import numpy as np
```

```python
character = pd.read_csv("character.csv")
character
```

```python
character_li = character.drop_duplicates().values.tolist()
character_li = sum(character_li, [])
kw_character = ['character', 'characters', 'characterization', 'role', 'roles'] + character_li
```

```python
for i in range(0, 100):
    target = "after_slicing/%r.csv" %i
    df = pd.read_csv(target, header=None)
    df.columns = ["RV"]
    
    f_list = []
    character = []
    
    for j in range(len(df)):
        f_list.append(df.RV[j])
        f_list[j] = f_list[j].split(" ")
    
    for j in range(len(f_list)):
        for keywords in kw_character:
            for k in range(len(f_list[j])):
                if keywords == f_list[j][k]:
                    character.append(f_list[j])
                    
    for j in range(len(character)):
        character[j] = " ".join(character[j])
        
    character = pd.DataFrame(character)
    character.to_csv("feature_assign/character/%r.csv" %i, index=False)
```
