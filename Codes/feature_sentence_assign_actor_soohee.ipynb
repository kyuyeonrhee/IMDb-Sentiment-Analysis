```python
import pandas as pd
import numpy as np
```

```python
actor = pd.read_csv("actor.csv")
actor
```

```python
actor_li = actor.drop_duplicates().values.tolist()
actor_li = sum(actor_li, [])
kw_actor = ['actor', 'actors', 'actress', 'actresses', 'acting'] + actor_li
```

```python
for i in range(0, 100):
    target = "after_slicing/%r.csv" %i
    df = pd.read_csv(target, header=None)
    df.columns = ["RV"]
    
    f_list = []
    actor = []
    
    for j in range(len(df)):
        f_list.append(df.RV[j])
        f_list[j] = f_list[j].split(" ")
    
    for j in range(len(f_list)):
        for keywords in kw_actor:
            for k in range(len(f_list[j])):
                if keywords == f_list[j][k]:
                    actor.append(f_list[j])
                    
    for j in range(len(actor)):
        actor[j] = " ".join(actor[j])
        
    actor = pd.DataFrame(actor)
    actor.to_csv("actor/%r.csv" %i, index=False)
```
