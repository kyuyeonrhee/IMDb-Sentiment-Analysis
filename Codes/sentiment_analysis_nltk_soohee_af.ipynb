```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer 
import pandas as pd
```

```python
for i in range(0, 100):
    target = "feature_assign/screenplay/%r.csv" %i
    
    df = pd.read_csv(target)
    df.columns = ['RV']
    df = df.drop_duplicates()
    df = df.reset_index()
    df = df.drop(columns="index")
    
    rv_list = []

    for j in range(len(df)):
        rv_list.append(df.RV[j])
        
    sid = SentimentIntensityAnalyzer()
    for sentence in rv_list:
        #print(sentence)
        ss = sid.polarity_scores(sentence)
        #for k in ss:
        #    print('{0}: {1}, '.format(k, ss[k]), end='')
        #print()
        
    compound_list = []

    for sentence in rv_list:
        compound_list.append(sid.polarity_scores(sentence)['compound'])
        
    cl = pd.DataFrame({'col': compound_list})
    cl.to_csv("feature_assign/screenplay/compound/%r.csv" %i)
    
    compound = sum(compound_list)/len(df)
    print(compound)
```
