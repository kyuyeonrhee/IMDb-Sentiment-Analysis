## crawler


```python
def get_reviews(url):
    driver.get(url)
    
    # review 개수
    num_reviews = driver.find_element_by_css_selector('#main > section > div.lister > div.header > div > span')
    num_reviews = num_reviews.text.split(" ")[0]
    num_reviews = num_reviews.replace(",", "")
    num_reviews = int(num_reviews)
    
    # click 몇 번 할지 결정
    if int(num_reviews) % 25 == 0:
        end = int(num_reviews/25) - 1
    else:
        end = int(num_reviews/25)

    # 마지막 review 보일 때까지 click 반복
    for i in range(0, end):
        driver.find_element_by_xpath('//*[@id="load-more-trigger"]').click() 
        time.sleep(5)
    
    # crawl review content
    review_content = driver.find_elements_by_class_name("show-more__control")
    
    reviews = []
    
    # text 추출해서 list에 저장
    for i in range(len(review_content)):
        reviews.append(review_content[i].text)

    print("# of reviews: {}".format(len(reviews)))
    return(reviews)
```

```python
# review text를 각각 csv 파일로 저장
for i in range(0, len(url)):
    print(i)
    data = get_reviews(url["url"][i])
    review_data = pd.DataFrame(data)
    review_data.to_csv("%r.csv" %i)
```

```python
# review text를 각각 txt 파일로 저장
for i in range(0, 2):
    print(i)
    data = get_reviews(url["url"][i])
    review_data = pd.DataFrame(data)
    review_data.to_csv("%r.txt" %i)
```
